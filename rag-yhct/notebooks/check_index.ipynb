```python
import pickle
from pathlib import Path
from pprint import pprint

# Tu dong tim dung duong dan tu notebook hoac project root
candidates = [
    Path('data/indexes/bm25_chunks_v2_full.pkl'),
    Path('../data/indexes/bm25_chunks_v2_full.pkl'),
]
pkl_path = next((p for p in candidates if p.exists()), None)
if pkl_path is None:
    raise FileNotFoundError('Khong tim thay bm25_chunks_v2_full.pkl trong data/indexes')

with pkl_path.open('rb') as f:
    data = pickle.load(f)

print(f'Loaded from: {pkl_path.resolve()}')
print(f'Type: {type(data).__name__}')

# Hien thi thong tin tong quan + 5 mau dau
if isinstance(data, dict):
    print(f'Total keys: {len(data)}')
    first_items = list(data.items())[:5]
    for i, (k, v) in enumerate(first_items, start=1):
        print(f'\n[{i}] key = {k}')
        if isinstance(v, (list, tuple)) and len(v) > 3:
            print(f'value type: {type(v).__name__}, len={len(v)}, preview={v[:3]}')
        else:
            print(f'value type: {type(v).__name__}')
            pprint(v)
elif isinstance(data, (list, tuple)):
    print(f'Total items: {len(data)}')
    pprint(data[:5])
else:
    print('Object is not dict/list/tuple, printing direct preview:')
    pprint(data)
```
